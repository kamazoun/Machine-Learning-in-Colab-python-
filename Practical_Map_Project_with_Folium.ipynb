```
import folium
import pandas

map = folium.Map(location=(0, -0), zoom_start = 6, tiles = "Mapbox Bright")

#Use cases, inefficient to load multiple values of map markers
#map.add_child(folium.Marker(location=[38.2, -99.1], popup="Hi! I'm Arkansas", icon = folium.Icon(icon="far fa-circle", color="green", prefix="fa")))
#fg = folium.FeatureGroup(name="My map")
#fg.add_child(folium.Marker(location=[38.2, -99.1], popup="Hi! I'm Kansas", icon = folium.Icon(color="green")))

fgv = folium.FeatureGroup(name="Volcanoes")
fgp = folium.FeatureGroup(name="Population")
fgc = folium.FeatureGroup(name="Cities")
data = pandas.read_csv("Volcanoes_USA.txt")

#print(data)

long = list(data["LON"])
lat = list(data["LAT"])
elev = list(data["ELEV"])

def altitude_color(elev):
  if elev < 1000:
    return "green"
  elif elev < 3000:
    return "orange"
  else:
    return "red"


for ln, lt, el in zip(long, lat, elev):
  #print(lt, ln, el)
  fgv.add_child(folium.CircleMarker(location=[lt, ln], popup=folium.Popup(str(el) + " m"), radius = 17, fill_color=altitude_color(el), fill_opacity=0.9, color="grey"))

fgp.add_child(folium.GeoJson(data=open("world.json", "r", encoding="utf-8-sig"), 
                           style_function=lambda x: {"fillColor":"yellow" if x["properties"]["POP2005"] < 10000000 
                                                     else 'orange' if 10000000 < x["properties"]["POP2005"] < 20000000 else 'red'}))

fgc.add_child(folium.Marker(location=[38.2, -99.1], popup="Hi! I'm Kansas", icon = folium.Icon(color="blue")))

map.add_child(fgc)
map.add_child(fgp)
map.add_child(fgv)
map.add_child(folium.LayerControl())

map.save("Map1.html")

map
```
